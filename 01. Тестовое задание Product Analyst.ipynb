# 3. Зачем нужен А/А тест? Опиши, как бы ты его проводил..

Чтобы убедиться, что та метрика, с которой мы работаем, не меняется в отсутствие воздействия. Данный тест производится на истории, до А/В теста.

Если данных мало(продукт новый), то можно производить А/А/В тест.


Описание проведения А/А теста:

А/А тесты необходимо проводить не один раз, а регулярно (в идеале).
Стандартный алгоритм:

- Определение гипотезы, метрик и критериев;

- Подбор групп;

- Проведение эксперимента;

- Оценка результатов:

     * Проверка на нормальности;
     * T test, U test
     * Определение разницы между группами (Lift)
     
- Подробное документирование эксперимента.
